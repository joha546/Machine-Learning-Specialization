# Data Generation:

Import libraries like numpy and matplotlib.pyplot for numerical computations and plotting.
Define a function to generate random data points. This function can create points following a specific linear relationship with added noise to simulate real-world scenarios.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_data(num_points, slope, bias, noise):
    x=np.random.rand(num_points)
    y=slope*x+bias+noise*np.random.rand(num_points)
    return x,y
```

# Model Definition:

Implement a simple linear regression model. This model will take an input feature (x) and predict an output (y). It will have weights (w) and a bias term (b).

```python
def predict(x,w,b):
    return w*x+b
```

# Loss Function:

Define a function to calculate the loss between predicted and actual values. Here, we'll use the Mean Squared Error (MSE).

```python
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)
```

# Empirical Risk Minimization

Implement the core ERM loop. Iterate through each data point and calculate the total loss for the entire training set. You can use a graidient descent approach to update weights and bias to minimize loss.

```python
loss_history=[]
def train(x,y,learning_rate,epochs):
    w=0
    b=0
    for epoch in range(epochs):
        y_pred=predict(x,w,b)
        loss=mse_loss(y,y_pred)
        w -=learning_rate*np.mean(x*(y_pred-y))
        b -= learning_rate * np.mean(y_pred - y)
        loss_history.append(loss) 
    return w,b
```

Data

```python
x,y=generate_data(100, 2, 3, 0.1)

w,b = train(x, y, 0.01, 300)

y_pred = predict(x, w, b)
```

```python
# Plot data and model
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Data and Learned Model')
plt.show()

# Plot loss vs epochs (assuming you store loss in each epoch)
plt.plot(loss_history)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss over Training')
plt.show()
```
